```python
fig = plt.figure(figsize=(12, 4))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)

ax1.plot(price.index, price['portfolio return'], linewidth = 2.0)
ax2.plot(price.index, price['volatility (20 days)'], linewidth = 2.0)

ax1.set_ylabel("Return")
ax2.set_ylabel('Volatility')
```

```python
return_risk = (price['portfolio return'] / price['volatility (20 days)']).ffill().values

fig = plt.figure(figsize = (12, 8))
plt.plot(price.index, return_risk)
print('Max:', np.nanmax(return_risk), 'Min:', np.nanmin(return_risk))
```
