```python
import pandas as pd
from pandas import DataFrame, Series
import datetime
import numpy as np
import matplotlib.pyplot as plt
```

```python
price = pd.read_excel(open('hw3_vol_ctrl_index1.xlsx', 'rb'), index_col = 0, sheet_name = 'initial_data')

price.head()
```

```python
price.iloc[0, 0]
```

```python
initial_capital = 1000000
initial_weights = [0.8, 0.06,0.04,0.06,0.04]

initial_position = []
for i in range(len(initial_weights)):
    invested = initial_capital* initial_weights[i]
    #initial_position.append(new)
    pos = invested / price.iloc[0, i]
    initial_position.append(pos)

initial_position
```

```python
index = {}
for time, p in price.iloc[:, :-3].iterrows():
    #print(time, p.values)
    index_value = initial_position * p.values
    index[time] = index_value
    
```

```python
portfolio_df = DataFrame.from_dict(index).T
portfolio_df.columns = price.columns[:-3]

portfolio_df.head()
```

```python
etf_value = [portfolio_df[i:i+1].values for i in range(len(portfolio_df))]
portfolio_df['index'] = [sum(value[0]) for value in etf_value]

portfolio_df.head()
```
